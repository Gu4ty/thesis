```python
import sys
sys.path.append("../")

import random

INTERVAL_LOW = 0
INTERVAL_HIGH = 100
SAMPLES = int(1e5)

a = 0.04
b = 0.0005
c = 0.2
d = 0.004
normalization_factor = 100

def system(y, t,a,b,c,d):
    X, Y = y   
    dXdt =  X*(a - b*Y) 
    dYdt = -Y*(c - d*X)
    return dXdt, dYdt

def get_time_norm_function(function, max_time):
    def function_norm_time(t):
        return function(t * max_time)

    return function_norm_time


def initial_conditions_lotka(population_size, number_of_initial_conditions):
    def generate_initial_conditions(population_size):
        N = population_size
        x = random.randint(1,N -1)
        y = N - x
        return  [x,y]
    
    yield [population_size//2,population_size//2]
    for i in range(number_of_initial_conditions -1):  
        yield generate_initial_conditions(population_size)

include_time = True
# The ideal trained neural network model
class ExactModel:
    def eval_model(self, X):
        if include_time:
            return [X[1]* normalization_factor * (a - b * X[2] * normalization_factor)]
        else:
            return [X[0]* normalization_factor * (a - b * X[1] * normalization_factor)]
#         if include_time:
#             return [-X[2]* normalization_factor * (c - d * X[1] * normalization_factor)]
#         else:
#             return [-X[1]* normalization_factor * (c - d * X[0] * normalization_factor)]

exact = ExactModel()
```

```python
from interpolation.interpolate_system import odeint_interpolate
from interpolation.regression_dataset import RegressionDataset

import matplotlib.pyplot as plt
import numpy as np


number_of_systems = 7
# samples_per_system = 14285
# samples_per_system = 1428
# number_of_systems = 1
samples_per_system = 14

dataset_train = None
dataset_test = None

population_size = 50

print("Generating training data...")
# f, axes = plt.subplots(2, 1, sharex='col', sharey='col', figsize=(7, 15))
f, axes = plt.subplots(1, 2)
# f, axes = plt.subplots(1, 1)

for init_conditions in initial_conditions_lotka(population_size, number_of_systems):
#     print(init_conditions)
    y_inter = odeint_interpolate( system, init_conditions, INTERVAL_LOW, INTERVAL_HIGH, 10000, (a,b,c,d))
    y_inter_time_norm = [get_time_norm_function(yi, INTERVAL_HIGH) for yi in y_inter]
    
    t = np.linspace(INTERVAL_LOW, INTERVAL_HIGH, 10000)
    X,Y = y_inter
#     Print para tesis
    x0= 50
    def find_roots(x,y):
        s = np.abs(np.diff(np.sign(y))).astype(bool)
        return x[:-1][s] + np.diff(x)[s]/(np.abs(y[1:][s]/y[:-1][s])+1)
#     axes.axvline(x=x0, linestyle=':', color='red')
    axes[0].axvline(x=x0, linestyle=':', color='red')
    axes[1].axvline(x=x0, linestyle=':', color='red')
    z = find_roots(np.array([X(ti) for ti in t]), t-x0)
    axes[0].plot(np.ones_like(z)*x0, z, marker="o", ls="", ms=4, color="limegreen")
#     axes.plot(np.ones_like(z)*x0, z, marker="o", ls="", ms=4, color="limegreen")
    z = find_roots(np.array([Y(ti) for ti in t]), t-x0)
    axes[1].plot(np.ones_like(z)*x0, z, marker="o", ls="", ms=4, color="limegreen")
    

#     End print para tesis
    axes[0].plot(t, [X(ti) for ti in t])
    axes[1].plot(t, [Y(ti) for ti in t])
#     axes.plot(t, [X(ti) for ti in t], label = "x(t)")
#     axes.plot(t, [Y(ti) for ti in t], label = "y(t)")
    
    print(init_conditions)
    
    def y_prime(t):
        return y_inter_time_norm[0](t) * (a - b * y_inter_time_norm[1](t))
#         return -y_inter_time_norm[1](t) * (c - d * y_inter_time_norm[0](t))
    
    d_train = RegressionDataset(y_inter_time_norm, y_prime, normalization_factor, samples_per_system * 80 // 100, include_time, 0,1)
    d_test = RegressionDataset(y_inter_time_norm, y_prime, normalization_factor, samples_per_system * 20 // 100, include_time,0, 1)
   
    if dataset_train:
        dataset_train.concat_dataset(d_train)
        dataset_test.concat_dataset(d_test)
    else:
        dataset_train = d_train
        dataset_test = d_test




print("Done")
axes[0].grid()
axes[1].grid()
axes[0].set_title("x(t)")
axes[1].set_title("y(t)")
# axes[0].grid()
# axes.grid()
# axes.set_title("x(t) para 7 condiciones iniciales")
# axes.legend()
# plt.savefig('lotka_volterra_example_two_initial_conditions.png')
plt.show()
```

```python
import numpy as np
from scipy import stats

x = np.array([5, 15, 25, 35, 45, 55])
y = np.array([29, 11, 3, 8, 29, 30])



plt.scatter(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y');
plt.show()
```

```python

slope, intercept, r, p, std_err = stats.linregress(x, y)
print(slope,intercept)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y');
plt.plot(x, mymodel)
plt.show() 
```

```python
mymodel = np.poly1d(np.polyfit(x, y, 2))
print(mymodel)
myline = np.linspace(1, 55, 100)
plt.scatter(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y');
plt.plot(myline, mymodel(myline))
plt.show() 
```
