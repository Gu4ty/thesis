```python
INTERVAL_LOW = 0
INTERVAL_HIGH = 100
SAMPLES = int(1e5)
N = 50
normalization_factor = N

def system(y, t):
    S, I, R = y
    dSdt = -0.2 * S * I / N
    dIdt = 0.2 * S * I / N - 0.1 * I
    dRdt = 0.1 * I
    return dSdt, dIdt, dRdt

def get_time_norm_function(function, max_time):
    def function_norm_time(t):
        return function(t * max_time)

    return function_norm_time



class ExactModel:
    def eval_model(self, X):
        return [-0.2 * (X[1] * normalization_factor) * X[2] * normalization_factor / N]
        


exact = ExactModel()
```

```python
from interpolation.regression_dataset import RegressionDataset
from interpolation.interpolate_system import odeint_interpolate

init_conditions = [49,1,0]
include_time = False

y_inter = odeint_interpolate( system, init_conditions, INTERVAL_LOW, INTERVAL_HIGH, 10000 )
y_inter_time_norm = [get_time_norm_function(yi, INTERVAL_HIGH) for yi in y_inter]
y_inter_time_norm = y_inter_time_norm[:2]

def y_prime1(t):
    return float(-0.2 * y_inter_time_norm[0](t) * y_inter_time_norm[1](t) / N)

print("Generating training data...")
dataset_train = RegressionDataset(y_inter_time_norm, y_prime1, normalization_factor, SAMPLES * 80 // 100, include_time)
dataset_test = RegressionDataset(y_inter_time_norm, y_prime1, normalization_factor, SAMPLES * 20 // 100, include_time)
print("Done")
```
