```python
INTERVAL_LOW = 0
INTERVAL_HIGH = 100
SAMPLES = int(1e5)

N = 50
normalization_factor = N

def system(y, t):
    S, I, R = y
    dSdt = -0.2 * S * I / N
    dIdt = 0.2 * S * I / N - 0.1 * I
    dRdt = 0.1 * I
    return dSdt, dIdt, dRdt

def get_time_norm_function(function, max_time):
    def function_norm_time(t):
        return function(t * max_time)

    return function_norm_time



class ExactModel:
    def eval_model(self, X):
        return [-0.2 * (X[1] * normalization_factor) * X[2] * normalization_factor / N]


exact = ExactModel()
```

```python
from interpolation.dataset_exact import ExactDataset
from interpolation.interpolate_system import odeint_interpolate

y_inter = odeint_interpolate( system, [49, 1, 0], INTERVAL_LOW, INTERVAL_HIGH, 10000 )
y_inter_time_norm = [get_time_norm_function(yi, INTERVAL_HIGH) for yi in y_inter]

def f(t, y):
    return (-0.2 * y[0] * y[1]/ N)

def y_prime(t):
    return (-0.2 * y_inter_time_norm[0](t) * y_inter_time_norm[1](t) / N)

print("Generating training data...")
dataset_train = ExactDataset(y_inter_time_norm, f,normalization_factor ,SAMPLES * 80 // 100)
dataset_test = ExactDataset(y_inter_time_norm, f,normalization_factor ,SAMPLES * 20 // 100)
print("Done")
```

```python
%load_ext autoreload
%autoreload 2
from interpolation.regression_nn import RegressionNN

m1 = RegressionNN.nn_regression(y_inter_time_norm, y_prime, int(1e5), 2048, 30, dataset_train, dataset_test,True)
```

```python
%load_ext autoreload
%autoreload 2
from query_set.graph_dependency import dependency_graph_epsilon
import matplotlib.pyplot as plt

f, axes = plt.subplots(5, 6, figsize=(30, 20))
for i in range(5):
    for j in range(6):
        dependency_graph_epsilon(
            m1,
            1,
            y_inter_time_norm,
            0.2,
            0.7,
            normalization_factor,
            5,
            1000,
            True,
            "",
            "b",
            "",
            "",
            exact,
            axes[i,j]
        )

f.supxlabel('S')
f.supylabel('N(S)')
f.suptitle("N(S) for t,I,R fixed. Color red is the exact ideal model", fontsize=14)
plt.savefig('sir_N_50_X_si.pdf', bbox_inches='tight')
plt.show()
```
