```python
INTERVAL_LOW = 0
INTERVAL_HIGH = 100
```

```python
import random
import numpy as np

b = 1/2
k = 1/3
normalization_factor = 1
def system(y, t,b,k):
    s, i, r = y
    dSdt = -b * s * i
    dIdt = b * s * i  - k * i
    dRdt = k * i
    return dSdt, dIdt, dRdt



def get_time_norm_function(function, max_time):
    def function_norm_time(t):
        return function(t * max_time)

    return function_norm_time

include_time = False
class ExactModel:
    def eval_model(self, X):
        if include_time:
            return [-b * X[1] * X[2]]
        else:
            return [-b * X[0] * X[1]]
         
def initial_conditions_sir(number_of_initial_conditions):
    def generate_initial_conditions(sum_of_numbers = 1):
        l = [random.uniform(0, sum_of_numbers) for i in range(2)]
        l+= [0, sum_of_numbers]
        l.sort()
        sir = []
        for i in range(1, len(l)):
            sir.append(l[i] - l[i-1])
        return sir
    
    yield [1,1.27 * 1e-6,0]
    for i in range(number_of_initial_conditions -1):  
        yield generate_initial_conditions()
        
exact = ExactModel()
```

```python
from interpolation.interpolate_system import odeint_interpolate
from interpolation.regression_dataset import RegressionDataset

import matplotlib.pyplot as plt

number_of_systems = 10
samples_per_system = 10000

dataset_train = None
dataset_test = None

print("Generating training data...")
f, axes = plt.subplots(3, 1, sharex='col', sharey='col', figsize=(7, 15))
for init_conditions in initial_conditions_sir(number_of_systems):
    assert sum(init_conditions) - 1 < 1e-4, f"sum is {sum(init_conditions)}, must be 1"
    y_inter = odeint_interpolate( system, init_conditions, INTERVAL_LOW, INTERVAL_HIGH, 10000, (b,k))
    y_inter_time_norm = [get_time_norm_function(yi, INTERVAL_HIGH) for yi in y_inter]
    t = np.linspace(INTERVAL_LOW, INTERVAL_HIGH, 10000)
    s,i,r = y_inter
    axes[0].plot(t, [s(ti) for ti in t])
    axes[1].plot(t, [i(ti) for ti in t])
    axes[2].plot(t, [r(ti) for ti in t])
    
    y_inter_time_norm = y_inter_time_norm[:2]
    
    print(init_conditions)
    def y_prime(t):
        return float(-b * y_inter_time_norm[0](t) * y_inter_time_norm[1](t))
    
    d_train = RegressionDataset(y_inter_time_norm, y_prime, normalization_factor, samples_per_system * 80 // 100, include_time)
    d_test = RegressionDataset(y_inter_time_norm, y_prime, normalization_factor, samples_per_system * 20 // 100, include_time)
   
    if dataset_train:
        dataset_train.concat_dataset(d_train)
        dataset_test.concat_dataset(d_test)
    else:
        dataset_train = d_train
        dataset_test = d_test

print("Done")
print(dataset_train.X[0])
print(dataset_train.Y[0])
axes[0].set_title("Diferentes S(t)")
axes[1].set_title("Diferentes I(t)")
axes[2].set_title("Diferentes R(t)")
  
plt.show()
    
```
